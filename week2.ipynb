## 3. Simple Statistics

```python
# Key scalar stats for Exam_Score
avg_score   = df["Exam_Score"].mean()
max_score   = df["Exam_Score"].max()
min_score   = df["Exam_Score"].min()
med_score   = df["Exam_Score"].median()
std_score   = df["Exam_Score"].std()

print("=" * 40)
print("  EXAM SCORE SUMMARY")
print("=" * 40)
print(f"  {'Average Score':<20}: {avg_score:.2f}")
print(f"  {'Highest Score':<20}: {max_score}")
print(f"  {'Lowest Score':<20}: {min_score}")
print(f"  {'Median Score':<20}: {med_score:.2f}")
print(f"  {'Std Deviation':<20}: {std_score:.2f}")
print(f"  {'Total Students':<20}: {len(df):,}")
```
